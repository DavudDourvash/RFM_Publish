```python
import nbformat
import re

path = "MonthlyRFM_Crypto.ipynb"

# فایل نوت‌بوک را بخوان
nb = nbformat.read(path, as_version=4)

# شمارنده‌ی سلول‌ها
counter = 1

for cell in nb.cells:
    if cell.cell_type == "code":
        # حذف شماره‌های قدیمی در خط اول اگر وجود دارند
        lines = cell.source.splitlines()
        if lines and re.match(r"#\s*\[\d+\]", lines[0]):
            lines = lines[1:]
        # اضافه‌کردن شماره‌ی جدید
        cell.source = f"# [{counter}] " + "\n" + "\n".join(lines)
        counter += 1

# فایل را ذخیره کن
nbformat.write(nb, path)
print(f"✅ سلول‌ها بازشماری شدند ({counter-1} سلول کد).")
```
